# Q3

```python
M = np.array([[10,0,0,0], [0,20,0,0], [0,0,5,0], [0,0,0,15]])
K = np.array([[4, -3, 0, 0], [-3, 8, -5, 0], [0, -5, 11, -6], [0, 0, -6, 8]])
eig = np.linalg.eig(np.matmul(K, np.linalg.inv(M)))
frq = np.sqrt(eig[0])
eig[0], frq
```
